```julia
using Plots
```

```julia
path = "../output/all_triangles_1.0";
dataset_name = "NDC-classes";
```

```julia
fullpath = "$path/$(dataset_name)-triangles.txt"
f = open(fullpath, "r")
weights = readlines(f)
weights = [parse(Int64, x) for x in weights]
close(f)
```

```julia
histogram(weights, bins=:auto, xlabel="triangle weight", ylabel="count", label="count",
          title="Distribution of triangle weights for $dataset_name.")
```
