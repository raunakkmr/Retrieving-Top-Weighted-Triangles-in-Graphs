```julia
using Plots
```

```julia
path = "../output/all_triangles_1.0";
dataset_name = "email-Eu";
```

```julia
fullpath = "$path/$(dataset_name)-triangles.txt"
f = open(fullpath, "r")
weights = readlines(f)
weights = [parse(Int64, x) for x in weights]
close(f)
```

```julia
histogram(weights, bins=:auto, xlabel="triangle weight", ylabel="count", label="count",
          title="Distribution of triangle weights for $dataset_name.")
```

```julia
fullpath = "$path/$(dataset_name)-degree-to-weight.txt"
f = open(fullpath, "r")
weights = readlines(f)
deg_to_weight = [split(x) for x in weights]
close(f)
degs = [parse(Float64, x[1]) for x in deg_to_weight];
weights = [parse(Float64, x[2]) for x in deg_to_weight];
```

```julia
scatter(degs, weights)
```
